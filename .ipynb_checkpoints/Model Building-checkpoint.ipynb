```python
df.iloc[0].cast
def Take3Names(obj):
    l=[]
    x=3
    for i in ast.literal_eval(obj):
            if (x>0):
                l.append(i["name"])
                x=x-1
            else:
                break
    return l    
    
```

```python
def TakeDirector(obj):
    l=[]
    for i in ast.literal_eval(obj):
            if i["job"]=="Director":
                l.append(i["name"])
    return l  
```

```python
df.info()
```

```python
df["crew"]=df["crew"].apply(TakeDirector)
```

```python
df["overview"]=df["overview"].apply(lambda x:x.split())
```

```python
df.head()
```

```python
df["genres"]=df["genres"].apply(lambda x:[i.replace(" ","") for i in x])
df["crew"]=df["crew"].apply(lambda x:[i.replace(" ","") for i in x])
df["cast"]=df["cast"].apply(lambda x:[i.replace(" ","") for i in x])
df["keywords"]=df["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.head()
```

```python
df["tags"]=df["overview"]+df["genres"]+df["keywords"]+df["cast"]+df["crew"]
```

```python
df=df.drop(columns=["overview","genres","keywords","cast","crew"])
```

```python
df["tags"]=df["tags"].apply(lambda x:" ".join(x))
```

```python
df["tags"][0]
```

```python
df["tags"]=df["tags"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vector=cv.fit_transform(df["tags"]).toarray()
```

```python
vector
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
angles=cosine_similarity(vector)
```

```python
sorted(list(enumerate(angles[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    index=df[df["title"]==movie].index[0]
    distances=angles[index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(df.iloc[i[0]].title)
    
```

```python
recommend('Step Up')
```
